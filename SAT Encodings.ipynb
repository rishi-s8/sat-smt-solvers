# Pigeon Hole Principle

```python
from z3 import *
import numpy as np
from time import time
```

```python
n = 50
h = n-1
```

```python
vs = [ [Bool("p_{}_{}".format(i,j)) for j in range(h)] for i in range(n)]
```

```python
print(vs)
```

```python
tic = time()
Constraints = []
C = True
for i in vs:
    C = And(C, Or(i))
Constraints.append(C)

tic2 = time()
C=True
for i in range(n):
    for j in range(i+1,n):
        for k in range(h):
            C = And(C, Not(And(vs[i][k], vs[j][k])))
Constraints.append(C)
print("Quadratic Time:", time()-tic2)
```

```python
def sumToOne(vs):
    atLeastOne = Or(vs)

    s =  [Bool('s_{}'.format(i)) for i in range(len(vs))]
    atMostOne = Implies(vs[0], s[0])
    for i in range(1,n):
        atMostOne = And(atMostOne, And(Implies(vs[i], s[i]), Implies(s[i-1], Not(vs[i]))))
    return And(atLeastOne, atMostOne)
```

```python
tic = time()
Constraints = []
C = True
for i in vs:
    C = And(C, Or(i))
Constraints.append(C)

tic2 = time()
C=True
for k in range(h):
    s =  [Bool('s_{}'.format(i)) for i in range(n)]   
    C = And(C, Implies(vs[0][k], s[0]))
    for i in range(1,n):
        C = And(C, And(Implies(vs[i][k], s[i]), Implies(s[i-1], Not(vs[i][k]))))
Constraints.append(C)
print("Linear Time:", time()-tic2)
```

```python
s = Solver()
s.add(And(Constraints))
r = s.check()
if r == sat:
    m = s.model()
    for i in vs:
        for j in i:
            if is_true(m[j]):
                print(j)
    print("sat")
else:
    print("unsat")
print("time: ", time()-tic)
```
